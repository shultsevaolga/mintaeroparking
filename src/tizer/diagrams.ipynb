```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
for terminal, color in ((1, "green"), (2, "red"), (3, "blue"), (5, "darkcyan"), (None, "darkblue")):
    timeline_df = pd.read_csv(f"../svo_hackaton/data/timeline_{terminal}.csv", index_col="times")
    timeline_df = timeline_df.loc[:,~(timeline_df==0).all(axis=0)]
    plt.figure(figsize=(20, 3))
    sns.heatmap(timeline_df.transpose(), cmap=["white", color], cbar=False)
    plt.show()
```
